```python
import numpy as np
import pandas as pd
```

```python
user_table = pd.read_table('./users.dat',index=True,header=None,names=['UserID','Gender','Age','Occupation','Zip-code'],sep='::',engine='python')
user_table
```

```python
rantings_table = pd.read_table('./ratings.dat',engine='python',sep='::',header=None,names=['UserID', 'MovieID', 'Rating', 'Timestamp'])
rantings_table
```

```python
movies_table = pd.read_table('movies.dat',sep='::',header=None,engine='python',names=['MovieID', 'Title', 'Genres'])
movies_table
```

```python
table = pd.merge(pd.merge(user_table,rantings_table),movies_table)
table.tail(5)
```

```python
table.info()
```

```python
table.describe()
```

```python
table[table['UserID']==1]
```

```python
#查看每一部电影不同性别的平均评分并计算分歧差值，之后排序
table_gender = table.pivot_table(index = ['Title'],columns='Gender',values='Rating',aggfunc='mean')
table_gender
```

```python
table_gender['diff']=np.fabs(table_gender['M']-table_gender['F'])
table_gender.head()
```

```python
sorted_table = table_gender.sort_values(by='diff',ascending=False)
sorted_table.head(10)
```

```python
#算出每部电影平均得分并对其进行排序
rant_table = table.pivot_table(index='Title',values='Rating').sort_values(by='Rating',ascending=False)
rant_table
```

```python
# 查看评分次数多的电影并进行排序 
count = pd.crosstab(table['Title'],table['Rating'],margins=True)
count
```

```python
count_table = count.sort_values(by='All',ascending=False)
count_table
```

```python
#过滤掉评分条目数不足250条的电影
count_table[count_table['All']>=250]
```

```python
# 评分最高的十部电影
acd = table.groupby('Title').agg({'Rating':[np.mean,np.size]})
acd.head(10)
```

```python
sort = acd['Rating']['size']>100
acd[sort]['Rating'].sort_values(by='mean',ascending=False).head(10)
```

```python
#查看不同年龄的分布情况并且采用直方图进行可视化
import matplotlib.pyplot as mp
```

```python

mp.xlabel('age',fontsize=14)
mp.ylabel('count',fontsize=14)
labels=['0-9','10-19','20-29','30-39','40-49','50-59','60-69','70-79']
mp.hist(table.Age,bins=30)
mp.show()
```

```python
labels=['0-9','10-19','20-29','30-39','40-49','50-59','60-69','70-79']
table['age_group']=pd.cut(table.Age,range(0,81,10),labels=labels)
table.head()
```

```python
table['age_group'].value_counts().plot(kind='bar')
mp.xticks(rotation=45)
mp.show()
```
